```python
import rdkit
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem.Draw import MolsToGridImage

print(sample_smiles)

sample_moles= [Chem.MolFromSmiles(smiles) for smiles in sample_smiles]
grid=rdkit.Chem.Draw.MolsToImage(sample_moles,molsPerRow=5,subImaSize=(600,600),legends=None)
MolsToGridImage(mols=sample_moles, molsPerRow=4)
```
